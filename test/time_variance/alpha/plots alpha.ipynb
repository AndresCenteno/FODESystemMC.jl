```julia
using Plots
using Statistics
using DelimitedFiles
using LaTeXStrings
```

```julia
xlab = "nx"
xvec = 1:5
xticks = 1:5
```

```julia
alpha_exp = readdlm("alpha_checks.csv");
```

```julia
alpha_exp = reshape(alpha_exp,9,9,:);
```

```julia
mean_alpha_exp = mean(alpha_exp,dims=3)
```

```julia
plot(range(0.1,0.9,9),mean_alpha_exp[2:5,:]',xlabel=L"\alpha",
ylabel="Mean variance",title="1e3 particles over 40 random experiments",labels=["u(T)" L"\partial A" L"\partial \alpha" L"\partial u_0"])
savefig("alpha/variance_alpha.png")
```

```julia
plot(range(0.1,0.9,9),mean_alpha_exp[2:6,:]',xlabel=L"\alpha",
ylabel="Mean variance",title="1e3 particles over 40 random experiments",labels=["u(T)" L"\partial A" L"\partial \alpha" L"\partial u_0" L"\partial T"])
savefig("alpha/variance_alpha_with_time.png")
```

```julia
plot(range(0.1,0.9,9),mean_alpha_exp[7:9,:]',xlabel=L"\alpha",
ylabel="Mean time (nanoseconds)",title="1e3 particles over 40 random experiments",labels=["0.95 quantile" "0.05 quantile" "Mean time"])
```

```julia
savefig("alpha/time.png")
```
